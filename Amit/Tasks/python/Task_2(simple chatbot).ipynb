```python
import random

class Chatbot:
    
    def __init__(self, name):
        self.name = name
        self.greetings = ["Hello there!", "Hi!", "Greetings!", "Good day!"]
        self.farewells = ["Goodbye!", "See you later!", "Have a nice day!", "Farewell!"]
        self.questions = ["How can I help you today?", "What would you like to talk about?", "Is there anything specific you need assistance with?"]
        self.default = ["I'm sorry, I didn't understand.", "Could you please rephrase that?"]


    def greet(self):
        """
        describtion of this function : provide a friendly greeting to User
        1)self is used to reference the instance of chatbot
        2)Random Greeting Selection: select randomly from self.greetings
        """
        print(f"Chatbot : {random.choice(self.greetings)}")


    def farewell(self):
        """
        describtion of this function : provide a farewell message to the user when the end of interactions
        1)self is used to reference the instance of chatbot
        2)Random farewell Selection: select randomly from self.farewells
        """
        print(f"Chatbot : {random.choice(self.farewells)}")


    def ask_question(self):
        """
        Description of this function: To prompt the user to provide input or continue the conversation
        1) self is used to reference the instance of chatbot
        2) Random question selection: select randomly from self.questions
        """
        print(f"Chatbot : {random.choice(self.questions)}")

    def default_message(self):
        """
        Description of this function: chatbot response to the user who donot understand 
        1) self is used to reference the instance of chatbot
        2) Random question selection: select randomly from self.default
        """

        print(f"Chatbot : {random.choice(self.default)}")



    def process_input(self, user_input):
        """
        Description of this function:  Implement your input processing logic hereon
        1) to understand user intent and provide appropriate responses
        parameter:username ................................
        1)will take user name from user
        """
        print("You said:", user_input)         #print(f"Your said : {user_input}")


    def start_conversation(self):
        """
        Description of this function: To create an interactive and engaging conversation by repeatedly asking questions,
        processing user input.
        1)This function is typically called to start a new conversation with the user.
        2)It provides the main structure for the chatbot's interaction with the user, 
        handling the question-and-answer process and managing the conversation flow.
        3) called the upper function in this function(start_conversation)
        
        """
        
        self.greet()
        while True:
            self.ask_question()
            user_input = input().lower()
            
            if user_input in ["quit", "exit"] or user_input == "self.farewells":
                self.farewell()
                break
            else :
                self.process_input(user_input)
                self.default_message()
```
